```r
solar.radiacion <- c(11.1,10.6,6.3,8.8,10.7,11.2,8.9,12.2)
```

```r
# Hallando la media
mean(solar.radiacion)
# Hallando la mediana
median(solar.radiacion)
#Hallando la varianza
var(solar.radiacion)
```

```r
# Agrego 10 al objeto solar.radiacion y lo guardo en la variable sr10
sr10 <- solar.radiacion + 10
sr10
```

```r
# Hallando la media
mean(sr10)
# Hallando la mediana
median(sr10)
#Hallando la varianza
var(sr10)
```

13.Genera aleatoriamente 1.000 mascotas, de las opciones **perro**, **gato**, **pollo** y **pez dorado**, con la misma probabilidad de que cada uno sea elegido. Muestra los primeros valores de la variable resultante y cuente el número de cada tipo de mascota.

```r
# Creamos la muestra 

muestra <- sample(c("perro","gato","pollo","pez dorado"), 1000, replace = T)

#Mostramos los primeros valores de la variable resultante
muestra[1:20] # primera forma
head(muestra,20) # segunda forma 
```

```r
# Contando el numero de cada tipo de mascota

contadores <- numeric(4) #Creamos un vector de tipo numeric

for(i in muestra){
    if(i == "perro"){
        contadores[1] <- contadores[1] + 1 #Contadores1 acumula el número de perros
    } else if(i == "gato"){
        contadores[2] <- contadores[2] + 1  # Contadores2 acumula el número de gatos
    } else if( i == "pollo"){
        contadores[3] <- contadores[3] + 1  # Contadores3 acumula el número de pollos
    } else 
        contadores[4] <- contadores[4] + 1  # Contadores4 acumula el número de pez dorado
}
# Mostramos las cantidades de cada mascota 
sprintf("Cantidad de perros: %d", contadores[1]) 
sprintf("Cantidad de gatos: %d", contadores[2])
sprintf("Cantidad de pollo: %d", contadores[3])
sprintf("Cantidad de pez dorado: %d", contadores[4])
```
